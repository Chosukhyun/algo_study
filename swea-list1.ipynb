## 4828. min max
__Q. N개의 양의 정수에서 가장 큰 수와 작은 수의 차이를 출력하기__<br>
-입력
  - 첫 줄에 테스트 케이스의 수 T가 주어진다(1 <= T <= 50)
  - 각 케이스의 첫 줄에 양수의 개수 N이 주어진다(5 <= N <= 1000)
  - 다음 줄에 N개의 양수 ai가 주어진다(1 <= ai <= 1000000)
- 출력
   각 줄마다  "#T"(T는 테스트 케이스 번호)를 출력한 뒤, 답을 출력한다


```python
T = int(input())


for i in range(T):
  N = int(input())
  ai = list(map(int, input().split()))
  result = max(ai) - min(ai)
  print(f'#{i + 1} {result}')  
```

```python
# 버블 정렬

T = int(input())

for i in range(T):
  N = int(input())
  ai = list(map(int, input().split()))
  for j in range(N-1, 0, -1):
    for k in range(j):
      if ai[k] > ai[k+1]:
        ai[k], ai[k+1] = ai[k+1], ai[k]

m_n = ai[-1] - ai[0]

print('#{}{}' .format(i+1, m_n))
```

- 출력

```python
#풀이 1
T = int(input())

for i in range(T):
  N = int(input())
  ai = list(map(int, input()))
  c_list = [0 for _ in range(10)]

  for j in ai:
    c_list[j] += 1
  
  p_card = max(c_list)
  
  for j in ai:
    if c_list[j] == p_card:
      m_card = j

  print(f'#{i + 1} {m_card} {p_card}')

  
```

## 4835.구간합
- N개의 정수가 들어있는 배열에서 이웃한 M개의 합을 계산하는 것은 디지털 필터링의 기초연산이다.

- M개의 합이 가장 큰 경우와 가장 작은 경우의 차이를 출력하는 프로그램을 작성하시오.
 

- 다음은 N=5, M=3이고 5개의 숫자 1 2 3 4 5가 배열 v에 들어있는 경우이다.

```python
#풀이 1
T = int(input())

for i in range(T):
  N, M = map(int, input().split())
  ai = input().split()
  num = 0
  num_list = []

  for j in range(N-M+1):
    for k in range(j, j+M):
      num += int(ai[k])
    num_list.append(num)
    num = 0

ans = max(num_list) - min(num_list)

print(f'#{i+1} {ans}')
```
